```python
class Solution:
    def minimumAbsDifference(self, arr: [int]) -> [[int]]:
        arr.sort()
        lis = list()
        min = 2 ** 32
        i = 0
        while i < len(arr[:-1]):
            d = arr[i+1] - arr[i]
            if d == min:
                lis.append([arr[i], arr[i+1]])
            elif d < min:
                min = d
                lis = list()
                lis.append([arr[i], arr[i+1]])
            i += 1
        return lis
```

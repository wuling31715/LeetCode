```python
class Solution:
    def validPalindrome(self, s: str) -> bool:
        if s == s[::-1]:
            return True            
        j = 0        
        while j < len(s):
            b = s
            b = b[:j] + b[j+1:]
            if b == b[::-1]:
                return True            
            j += 1
        return False
```

```python
class Solution:
    def remove(self, a, b):
        flag = True
        i = 0
        while i < len(a[:-1]):
            if flag:                
                if a[i] != b[i]:
                    b = b[:i] + b[i+1:]
                    a = a[:len(a)-i] + a[len(a):]
                    flag = False
            else:
                if a[i] != b[i]:
                    return False
            i += 1
        return True
    def validPalindrome(self, s: str) -> bool:
        a = s
        b = s[::-1]
        right = self.remove(self, a, b)
        if right:
            return True
        left = self.remove(self, b, a)
        if left:
            return True        
        return False            
```

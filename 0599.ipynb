```python
class Solution:
    def findRestaurant(self, list1: [str], list2: [str]) -> [str]:
        dic = dict()
        dic2 = dict()
        i = 0
        while i < len(list1):
            dic.update({list1[i]: i})
            i += 1
        j = 0
        while j < len(list2):
            if list2[j] in dic:
                dic2.update({list2[j]: j + dic[list2[j]]})
            j += 1
        lis = [0]
        min = 2 ** 31
        for i, j in dic2.items():
            if j < min:
                lis.pop()
                lis.append(i)
                min = j
            elif j == min:
                lis.append(i)
        return lis
```

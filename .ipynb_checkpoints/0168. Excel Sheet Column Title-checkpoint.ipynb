```python
class Solution:
    def convertToTitle(self, n: int) -> str:
        alpha = 'ZABCDEFGHIJKLMNOPQRSTUVWXY'
        lis = [i for i in alpha]
        dic = {i: v for i, v in enumerate(lis)}
        s = ''
        while n > 0:
            tmp = n % 26
            if tmp == 0: # Z                 
                s += dic[tmp]
                n = n // 26 - 1
            else:
                s += dic[tmp]
                n = n // 26
        return s[::-1]
```

```python
s = Solution
```

```python
s.convertToTitle(s, 1)
```

```python
s.convertToTitle(s, 27)
```

```python
s.convertToTitle(s, 28)
```

```python
s.convertToTitle(s, 52)
```

```python
s.convertToTitle(s, 701)
```

```python
for i in range(1000):
    print(i, s.convertToTitle(s, i))
```
